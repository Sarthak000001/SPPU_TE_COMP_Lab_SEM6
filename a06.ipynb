```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='white')
```

```python
df = pd.read_csv("Iris.csv")
df

#for dataset without column name used below method
# column_names = ['PetalWidthCm','PetalWidthCm','PetalWidthCm','PetalWidthCm','Species']
# df = pd.read_csv("Iris.csv", names=column_names)
```

```python
print("Numeric features : SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm")
print("Nominal features : Species")
```

```python
# Visulatization of feature distributions using histogram
fig,axs = plt.subplots(2,2,figsize=(20,20))

sns.histplot(x='SepalLengthCm',hue='Species',data=df,ax=axs[0][0])
axs[0][0].set_title("Distribution of Sepal Length by Species")

sns.histplot(x='SepalWidthCm',hue='Species',data=df,ax=axs[0][1])
axs[0][1].set_title("Distribution of Sepal Width by Species")

sns.histplot(x='PetalLengthCm',hue='Species',data=df,ax=axs[1][0])
axs[1][0].set_title("Distribution of Petal Length by Species")

sns.histplot(x='PetalWidthCm',hue='Species',data=df,ax=axs[1][1])
axs[1][1].set_title("Distribution of Petal Width by Species")

plt.tight_layout()
plt.show()
```
