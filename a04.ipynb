```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
#Count of passengers survived or not survived
plt.figure(figsize=(12,12))
sns.countplot(x='survived',hue='survived',data=titanic)
plt.show()
```

```python
# count of passengers survived classwise
plt.figure(figsize=(12,12))
sns.countplot(x='class',hue='survived',data=titanic)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='class', y='age', data=titanic, palette='viridis')
plt.title('Age Distribution in Each Class')
plt.show()
```

```python
# count of survival of passengers by class and age
plt.figure(figsize=(12,12))
sns.boxplot(x='class',y='age',hue='survived',data=titanic)
plt.show()
```

```python
# Countplot based on gender
plt.figure(figsize=(12,12))
sns.countplot(x='class',hue='sex',data=titanic)
plt.show()
```

```python
# Survival status based on gender and age
plt.figure(figsize=(12,12))
sns.boxplot(x='sex',y='age',hue='survived',data=titanic)
plt.show()
```

```python
# checking how the price of the ticket (column name: 'fare')
# for each passenger is distributed by plotting a histogram
plt.figure(figsize=(10,6))
sns.histplot(titanic['fare'],bins=30,kde=True,color='red')
plt.title("Distribution of Ticket Prices")
plt.xlabel("Fare") #fare ->ticket price
plt.ylabel("Frequency")
plt.show()
```

Histogram conclusion :
Most fares were concentrated at lower values, with a long tail of higher fares. This suggests a diverse range of passengers with varying economic means, with the majority paying lower fares.
